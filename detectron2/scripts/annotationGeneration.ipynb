```python
from pycocotools import mask as cocoMask
def my_dataset_function():
    path = "/Users/mbronars/Documents/sohn/cellSegmentation/detectron2/rawImages/jpg/"
    temp = os.listdir("../pngMasks")
    annotations = []
    for f in temp:
        if ".png" != f[-4:]:
            continue
        #print(f)
        dic = {"file_name":path + f}
        dic["height"] = 308
        dic["width"] = 308
        img = cv.imread("../pngMasks/" + f)
        img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        print(img.shape)
        print(img)
        dic["annotations"] = []
        for c in range(1, 256):
            anndic = {}
            anndic["category_id"] = 1
            #anndic["bbox_mode"] = detectron2.structures.BoxMode.XYXY_ABS
            
            a = np.where(img != 0)
            bbox = np.min(a[0]), np.max(a[0]), np.min(a[1]), np.max(a[1])

            #anndic[""]
            mask = (img == [c])
            if mask.any():
                print("pixels of color: " + str(c))
                mask = mask.astype(np.uint8)
                ann = cocoMask.encode(np.asarray(mask, order="F"))
                dic["annotations"].append(anndic)
                a = np.where(mask != 0)
                bbox = np.min(a[0]), np.max(a[0]), np.min(a[1]), np.max(a[1])
                anndic["bbox"] = bbox
        annotations.append(dic)
    return annotations
```
